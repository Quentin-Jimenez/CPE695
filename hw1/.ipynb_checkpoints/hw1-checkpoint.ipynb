```python
import matplotlib.pyplot as plt 
plt.style.use('seaborn-whitegrid') 
import numpy as np 
noise_scale = 100 
number_of_samples = 50 
x = np.array(25*(np.random.rand(number_of_samples) - 0.8))
x.sort()
y = np.array(7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples) )
plt.plot(x,y,'ro')
```

```python
deg = 3
fit = np.polyfit(x,y,deg)
yfit = []

for i in range(len(x)):
    hold = 0
    for n in range(len(fit)):
        hold += fit[n] * (x[i]**(len(fit)-1-n))
    yfit += [hold]
    
plt.plot(x,yfit,x,y,'ro')
```

```python
deg_limit = 9
mse = []
dg = []

for j in range(deg_limit):
    fit = np.polyfit(x,y,j)
    y1 = []
    for i in range(len(x)):
        hold = 0
        for n in range(len(fit)):
            hold += fit[n] * (x[i]**(len(fit)-1-n))
        y1 += [hold]
    mse = mse + [np.square(np.subtract(y1, y)).mean()]
    dg += [j]

plt.plot(dg,mse)
plt.title("MSE vs Polynomial Degree")
```

As the degree increases, the MSE decreases, but reaches an asymptote at 3. While using a higher polynomial will reduce the error, a 3rd degree is good enough, and is the best choice for m.

```python
ns = [150, 200, 400, 600, 1000]
fig, axn = plt.subplots(len(ns), sharex=True, figsize=(6, 18))

for i in range(len(ns)):
    
    y_noise_fit = []
    
    for j in range(number_of_samples):
        y_noise = np.array(7 * x + 15 * x**2 + 1 * x**3 + ns[i]*np.random.randn(number_of_samples))
        
    fit_noise = np.polyfit(x,y_noise,deg)
    
    for k in range(number_of_samples):
        hold = 0
        for n in range(len(fit_noise)):
            hold += fit_noise[n] * (x[k]**(len(fit_noise)-1-n))
        y_noise_fit = y_noise_fit + [hold]
    
    axn[i].plot(x, y_noise,'ro')
    axn[i].plot(x, y_noise_fit)
    axn[i].set_title("Noise = " + str(ns[i]))
```

As the noise increases, the accuracy of the polynomial decreases. This is because the data becomes less like the polynomial it is generated from. Increasing the degree would not increase the accuracy as the base polynomial has not changed.
